```python
# Cria a lista rows contendo as colunas dos textos dos arquivos .txt
import csv

rows = []

with open(CORPUS_CSV, "r", encoding="utf-8") as f_csv: 
    df_sent = csv.reader(f_csv, delimiter=',', lineterminator='\n')  # lineterminator='\n'
             
    for row in df_sent:
        row = ''.join(row).replace('\n', ' ') #print(row)
        rows.append(row)
        
f_csv.close()  
print(len(rows))
```

```python
# Funcao para extrair as seções dispositivos da lista rows obtida anteriormente
#import pandas as pd
#special_chars = ['\t', '\n', '\r', '\x0b', '\x0c']
import re


def extract_pattern(text, pattern):
    
    #pattern = "(DISPOSITIVO).*"
    
    try:
        re.compile(pattern)
    except re.error:
        print("Padrao nao valido")
        exit()
    
    if (re.findall(pattern, str(text))):
        text_disp = re.search(pattern, str(text)).group()
        text_disp = str(text_disp).strip('\n')
    else:
        text_disp = "NA"
        
    return(text_disp)
```

```python
import pandas as pd

CORPUS2_CSV = PDF_FOLDER + 'DatasetDispSentencas.CSV'

disp = ["NA"]*len(rows)

def process_disp2():      
    
    pattern_disp = "(DISPOSITIVO).*"
    # Processa Regex e Extrai as seções Dispositivos de toda lista do Texto (rows) criando outra lista (disp)
    for i in range(len(rows)):
        disp[i] = extract_pattern(rows[i], pattern_disp)
    
    
    
    data = {
        "Texto": rows,
        "Dispositivo": disp
    }
    
    df_disp_NA = pd.DataFrame(data)
    
    df_disp_NA = df_disp_NA[df_disp_NA['Dispositivo'] != 'NA'] 
    
    print(len(df_disp_NA))
    
    df_disp_NA.to_csv(CORPUS2_CSV)
    
```
